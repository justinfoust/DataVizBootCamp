### Day 1 - Activities - 06-Stu_DownToInput

```python
myName = input("What is my name?")
yourName = input("What is the name of your neighbor?")
myCodingMonths = input("How many months have I been coding?")
yourCodingMonths = input("How many months has your neighbor been coding?")

print(f"I have been coding {myCodingMonths} months, {yourName} has been coding for {yourCodingMonths} months and together we have been coding for {int(myCodingMonths) + int(yourCodingMonths)}.")
```
