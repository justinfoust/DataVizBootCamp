```python
# The list of candies to print to the screen
candyList = ["Snickers", "Kit Kat", "Sour Patch Kids", "Juicy Fruit", "Swedish Fish",
             "Skittles", "Hershey Bar", "Starbursts", "M&Ms"]

# The amount of candy the user will be allowed to choose
allowance = 5

# The list used to store all of the candies selected inside of
candyCart = []

# Print out options
for i in range(len(candyList)):
    print("[" + str(i) + "] " + candyList[i])
    
# Print number of allowance
for i in range(allowance):
    candyCart.append(candyList[int(input('enter candy index'))])
    
for j in range(len(candyCart)):
    print(candyCart[j])
```

```python
# Print introduction
print('Welcome to the House of Pies! Here are our pies:\n\
\n\
-----------------------------------------------------------------------------------------')

# Create a list of types of pies available
pieOptions = ['Pecan',0],['Apple Crisp',0],['Bean',0],['Banoffee',0],['Blueberry',0],['Buko',0],['Burek',0],['Tamale',0],['Steak',0]

# Print list of pies horizontally with choice number
for pieIndex in range(len(pieOptions)): # for <var> in range<int>  -- This will cycle through by value
    
    if pieIndex != len(pieOptions) - 1:
        
        #'end' in print will append to the end of print value; default is new line
        print('('+str(pieIndex + 1)+') '+pieOptions[pieIndex][0], end=', ')
        
    else:
        
        print('('+str(pieIndex)+') '+pieOptions[pieIndex][0])
        
pieCount = 0
        
while True:
        
    #Prompt user to select a pie
    userChoice = int(input('Please enter the corresponding number of the pie you want.\n'))

    #Print response
    print(f'Great! We\'ll have that {pieOptions[userChoice - 1][0]} right out for you.')
    
    #Incriment pie count
    pieOptions[userChoice-1][1] += 1
        
    #Give choice to select another
    morePie = input('Would you like another pie? (Y)es or (N)o\n')
    
    #End loop if not yes
    if morePie != 'Y':
        break

print(f'You purchased:')

for pieIndex in range(len(pieOptions)):
    print(f'{pieOptions[pieIndex][1]} {pieOptions[pieIndex][0]}')
```
