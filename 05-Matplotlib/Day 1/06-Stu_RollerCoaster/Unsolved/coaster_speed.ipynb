```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dangerDrop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
railGun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
railGun
```

```python
time = np.arange(0, 130, 10)
time
```

```python
plt.title('Rollercoaster Speeds')
plt.xlabel('Time (sec)')
plt.ylabel('Speed (mph)')
plt.grid()

plt.plot(time, dangerDrop, 'r-', label='Danger Drop')
plt.plot(time, railGun, 'g-', label='Rail Gun')

plt.legend()
plt.show()
```
