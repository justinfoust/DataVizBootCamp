```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
temp_F = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
temp_C = [(i - 32) * 0.56 for i in temp_F]
```

```python
yrs = np.arange(1,13,1)
yrs
```

```python
plt.plot(yrs, temp_F)
plt.show()
```

```python
plt.plot(yrs, temp_C)
plt.show()
```

```python
plt.xlabel('Months')
plt.ylabel('Degrees')

plt.plot(yrs, temp_F, 'b+-', label='F')
plt.plot(yrs, temp_C, 'rs-', label='C')
plt.legend(loc=10)
plt.show()
```

```python
plt.plot(yrs, temp_F, 'b+-', yrs, temp_C, 'rs-')
```
