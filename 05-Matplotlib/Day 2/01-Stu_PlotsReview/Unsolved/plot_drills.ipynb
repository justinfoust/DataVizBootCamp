```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATASET 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
plt.bar(gyms,members)
plt.show()
```

```python
# DATASET 2
x_lim = 2 * np.pi
x_axis = np.arange(0, x_lim, 0.1)
sin = np.sin(x_axis)
```

```python
plt.plot(x_axis,sin)
plt.title('Sin(x)')
plt.xlabel('Real Numbers from 0 to 2$\pi$')
plt.ylabel('sin(x)')
plt.show()
```

```python
# DATASET 3
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
x_axis = np.arange(0, len(gyms))
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue"]
explode = (0, 0.05, 0, 0)
```

```python
plt.pie(members,labels=gyms, colors=colors,explode=explode)
plt.show()
```

```python
# DATASET 4
x_axis = np.arange(0, 10, 0.1)
times = []
for x in x_axis:
    times.append(x * x + np.random.randint(0, np.ceil(max(x_axis))))
```

```python
plt.scatter(x_axis,times)
plt.show()
```
