# Winning Wrestlers Entertainment

In this activity you will be taking four seperate csvs that were scraped down from a wrestling database, merging them together, and then creating charts to visualize a wrestler's wins and losses over the course of four years.

### Part 1 - Macho Merging

* You will likely need to perform three different merges over the course of this activity, changing the names of your columns as you go along.

```python
# Import the necessary modules
import os
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
```

```python
# Bring each CSV into a separate data frame
stats_2013 = pd.read_csv('../Resources/WWE-Data-2013.csv')
stats_2014 = pd.read_csv('../Resources/WWE-Data-2014.csv')
stats_2015 = pd.read_csv('../Resources/WWE-Data-2015.csv')
stats_2016 = pd.read_csv('../Resources/WWE-Data-2016.csv')

print(stats_2013.info())
print(stats_2014.info())
```

```python
# Merge the first two datasets (2013 and 2014) on "Wrestler" so that no data is lost (should be 182 rows)
stats_2013_2014 = pd.merge(stats_2013, stats_2014, on='Wrestler', how='outer')

stats_2013_2014.head(10)
```

```python
# Rename our _x columns to "2013 Wins", "2013 Losses", and "2013 Draws"

# Rename our _y columns to "2014 Wins", "2014 Losses", and "2014 Draws"
stats_2013_2014 = stats_2013_2014.rename(columns={
    'Wins_x':'Wins 2013',
    'Losses_x':'Losses 2013',
    'Draws_x':'Draws 2013',
    'Wins_y':'Wins 2014',
    'Losses_y':'Losses 2014',
    'Draws_y':'Draws 2014'
})

stats_2013_2014.head(10)
```

```python
# Merge our newly combined dataframe with the 2015 dataframe
stats_2013_2014_2015 = pd.merge(stats_2013_2014,stats_2015,on='Wrestler',how='outer')
stats_2013_2014_2015
```

```python
# Rename "wins", "losses", and "draws" to "2015 Wins", "2015 Losses", and "2015 Draws"
stats_2013_2014_2015 = stats_2013_2014_2015.rename(columns={
    'Wins':'Wins 2015',
    'Losses':'Losses 2015',
    'Draws':'Draws 2015'
})

stats_2013_2014_2015.head(10)
```

```python
# Merge our newly combined dataframe with the 2016 dataframe
stats_complete = pd.merge(stats_2013_2014_2015,stats_2016,on='Wrestler',how='outer')
stats_complete
```

```python
# Rename "wins", "losses", and "draws" to "2016 Wins", "2016 Losses", and "2016 Draws"

stats_comlete = stats_complete.rename(columns={
    'Wins':'Wins 2016',
    'Losses':'Losses 2016',
    'Draws':'Draws 2016'
})

stats_complete.head(10)
```
