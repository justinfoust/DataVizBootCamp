## 3. Show rows 0-30 only even rows and only columns that contain the word `Code`

```python
df3 = pd.read_csv("Data/2016-FCC-New-Coders-Survey-Data.csv")

#hint: use DataFrame.columns.values to get a list of all the columns available since there are 113 columns
df3 = df3.filter(regex='Code')

df3 = df3[(df3.index % 2 == 0) & (df3.index < 31)]

df3
```
