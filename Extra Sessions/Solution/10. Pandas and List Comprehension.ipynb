## 2. Filtering Data using Dates: Show only 2016 for both Feb and March

```python
df2 = pd.read_csv("../Data/dash_price.csv")
df2.info()
```

```python
# hint: Check if Datetime data type exists. If not, cast the field to Datetime
df2.Date = pd.to_datetime(df2.Date)
```

```python
# Option 1 using Masking Technique
mask1 = df2.Date.dt.year == 2016
mask2 = (df2.Date.dt.month == 2) | (df2.Date.dt.month == 3)
df2[mask1 & mask2]
```

```python
# Option 2 - Another way. Notice the use of ( ) brackets 
df2[(df2.Date.dt.year == 2016) & ((df2.Date.dt.month == 2) | (df2.Date.dt.month == 3))]
```

```python
# Option 3 with List Comprehension
df2[(df2.Date.dt.year == 2016) & [i in [2,3] for i in df2.Date.dt.month]]
```

## 3. Show rows 0-30 only even rows and only columns that contain the word `Code` 
* Demo advanced use of List Comprehension

```python
df3 = pd.read_csv("../Data/2016-FCC-New-Coders-Survey-Data.csv")

#hint: use .columns to get a list of all the columns available since there are 113 columns
df3.columns.values
```

```python
# first list comprehension [i for i in range(30) if i % 2 ==0] is for the ROWS
# second list comprehension ['Code' in i for i in df3.columns.values] for the COLUMNS
# remember .loc[ROW, COLUMN]

df3.loc[[i for i in range(30) if i % 2 ==0], ['Code' in i for i in df3.columns.values]]
```
