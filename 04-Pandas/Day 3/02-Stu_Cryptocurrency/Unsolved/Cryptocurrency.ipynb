```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
merged = pd.merge(dash_df,bitcoin_df,on='Date')

merged.head()
```

```python
# Rename columns so that they are differentiated
merged_renamed = merged.rename(columns={
    'Open_x':'Open Dash',
    'Hish_x':'High Dash',
    'Low_x':'Low Dash',
    'Close_x':'Close_Dash',
    'Volume_x':'Volume Dash',
    'Market Cap_x':'Market Cap Dash',
    'Open_y':'Open bitcoint'
})

merged_renamed.head()
```

```python
# alternatively you can set your suffixes when the merge occurs
merged1 = pd.merge(dash_df,bitcoin_df,on='Date',suffixes=(' Dash',' bitcoin'))

merged1.head()
```
