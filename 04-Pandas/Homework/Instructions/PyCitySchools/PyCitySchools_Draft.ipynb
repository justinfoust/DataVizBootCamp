```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_summary = pd.DataFrame({
    'School Type': school_data_complete.groupby(['school_name'])['type'].agg(lambda x:x.unique()),
    'Total Students': school_data_complete.groupby(['school_name'])['size'].agg(lambda x: x.unique()),
    'Total School Budget': school_data_complete.groupby(['school_name'])['budget'].agg(lambda x: x.value_counts().index[0]),
    'Per Student Budget': school_data_complete.groupby(['school_name'])['budget'].agg(lambda x: x.value_counts().index[0]) / school_data_complete.groupby(['school_name'])['size'].agg(lambda x: x.value_counts().index[0]),
    'Average Math Score': school_data_complete.groupby(['school_name'])['math_score'].mean(),
    'Average Reading Score': school_data_complete.groupby(['school_name'])['reading_score'].mean(),
    '% Passing Math': school_data_complete[school_data_complete['math_score'] > 69].groupby(['school_name'])['math_score'].count() / school_data_complete.groupby(['school_name'])['size'].agg(lambda x: x.value_counts().index[0]) * 100,
    '% Passing Reading': school_data_complete[school_data_complete['reading_score'] > 69].groupby(['school_name'])['reading_score'].count() / school_data_complete.groupby(['school_name'])['size'].agg(lambda x: x.value_counts().index[0]) * 100,
    '% Overall Passing Rate': school_data_complete[(school_data_complete['math_score'] + school_data_complete['reading_score']) / 2 > 69].groupby(['school_name'])['student_name'].count() / school_data_complete.groupby(['school_name'])['size'].agg(lambda x: x.value_counts().index[0]) * 100
})

school_summary_formatted = school_summary.copy()

school_summary_formatted['Total School Budget'] = school_summary_formatted['Total School Budget'].map('${:,.2f}'.format)
school_summary_formatted['Per Student Budget'] = school_summary_formatted['Per Student Budget'].map('${:,.2f}'.format)
school_summary_formatted['Average Math Score'] = school_summary_formatted['Average Math Score'].map('{:.3f}'.format)
school_summary_formatted['Average Reading Score'] = school_summary_formatted['Average Reading Score'].map('{:.3f}'.format)
school_summary_formatted['% Passing Math'] = school_summary_formatted['% Passing Math'].map('{:.3f}%'.format)
school_summary_formatted['% Passing Reading'] = school_summary_formatted['% Passing Reading'].map('{:.3f}%'.format)
school_summary_formatted['% Overall Passing Rate'] = school_summary_formatted['% Overall Passing Rate'].map('{:.3f}%'.format)

school_summary_formatted = school_summary_formatted.rename_axis('')

school_summary_formatted.sort_values(by=['% Overall Passing Rate'], ascending=False).head()
```

### Score by School Spending

```python
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

type(school_summary.iloc[1,3])
```

```python
school_summary['Spending Ranges (Per Student)'] = pd.cut(school_summary['Per Student Budget'],spending_bins,labels=group_names)

Score_by_Spending = pd.DataFrame({
    'Average Math Score': school_summary.groupby(['Spending Ranges (Per Student)'])['Average Math Score'].mean(),
    'Average Reading Score': school_summary.groupby(['Spending Ranges (Per Student)'])['Average Reading Score'].mean(),
    '% Passing Math': school_summary.groupby(['Spending Ranges (Per Student)'])['% Passing Math'].mean(),
    '% Passing Reading': school_summary.groupby(['Spending Ranges (Per Student)'])['% Passing Reading'].mean(),
    '% Overall Passing Rate': school_summary.groupby(['Spending Ranges (Per Student)'])['% Overall Passing Rate'].mean()
})

Score_by_Spending.applymap('{:.2f}'.format)
```

### Score by School Size

```python
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```
